## Beta

```python
def calculate_beta_new(ticker: str, stock_data_path: str) -> dict:
    """
    Calculate yearly beta values for a stock against the S&P 500 (^GSPC)
    
    Args:
        ticker: Stock ticker symbol (unused in current implementation but kept for consistency)
        stock_data_path: Path to the stock's CSV data file
    
    Returns:
        Dictionary with years as keys and beta values as values
    """
    # Download market data once
    market_data = yf.download("^GSPC", start="2020-01-01", end="2024-12-31")
    market_data.columns = market_data.columns.get_level_values(0)
    market_data['Market Return'] = market_data['Close'].pct_change()
    market_data.index = pd.to_datetime(market_data.index)
    
    try:
        # Load stock data
        stock_df = pd.read_csv(stock_data_path, skiprows=2)
        stock_df.rename(columns={stock_df.columns[0]: 'Date'}, inplace=True)
        stock_df['Date'] = pd.to_datetime(stock_df['Date'])
        stock_df = stock_df.sort_values('Date')
        stock_df.set_index('Date', inplace=True)
        stock_df['Stock Return'] = stock_df['Unnamed: 1'].pct_change()
        
        # Prepare market data
        market_df = market_data[['Close', 'Market Return']].copy()
        market_df = market_df.rename(columns={'Close': 'Market Close'})
        
        # Align both dataframes
        combined = stock_df.join(market_df, how='inner')
        combined = combined.dropna()
        
        # Initialize dictionary to store yearly betas
        yearly_betas = {}
        
        # Calculate beta for each year
        for year in range(2020, 2025):
            year_data = combined[combined.index.year == year]
            
            if len(year_data) > 1:  # Need at least 2 data points for covariance
                covariance = np.cov(year_data['Stock Return'], year_data['Market Return'])[0][1]
                variance = np.var(year_data['Market Return'])
                
                if variance != 0:  # Avoid division by zero
                    beta = covariance / variance
                    yearly_betas[year] = beta
                else:
                    yearly_betas[year] = np.nan
            else:
                yearly_betas[year] = np.nan
                
        return yearly_betas

    except Exception as e:
        print(f"Error processing {ticker}: {e}")
        return {year: np.nan for year in range(2020, 2025)}
```

#### Calculation of Expected Returns

```python
def calculate_expected_returns(betas, Rf, market_returns):
    expected_returns = []
    for i in range(len(betas)):
        if np.isnan(betas[i]):
            expected_returns.append(np.nan)
        else:
            # Convert Rf and market return to decimal
            rf_decimal = Rf[i] / 100
            market_return_decimal = market_returns[i] / 100
            
            er = rf_decimal + betas[i] * (market_return_decimal - rf_decimal)
            expected_returns.append(er)  # rounded to 4 decimals
    return expected_returns
```

#### Sharpe Ratio

```python
def calculate_sharpe_ratio(returns, risk_free_rate, risk):

    sharpe_ratios = []
    for i in range(len(returns)):
        try:
            r = float(returns[i])
            rf = float(risk_free_rate[i])
            sigma = float(risk[i])
            if np.isnan(r) or np.isnan(rf) or np.isnan(sigma):
                sharpe_ratios.append(np.nan)
            else:
                sr = (r - rf) / sigma
                sharpe_ratios.append(sr)
        except (ValueError, TypeError):
            sharpe_ratios.append(np.nan)
    return sharpe_ratios

def calculate_sharperatio(ticker):
    betas_dict = calculate_beta_new(ticker, f"/Users/tejasmacipad/Desktop/Extras/hackrush/Data/{ticker}.csv")
    betas = list(betas_dict.values())  # Convert dict to list if needed

    Rf = [0.89,1.45,2.95,3.88,4.20]

    exp_r = calculate_expected_returns(
        betas,
        Rf,
        annual[['Annual Return (%)']].values
    )
    

    print(exp_r)

    risk = final_data.loc[final_data["Symbol"] == ticker].iloc[:, 8:-1].values.flatten()
    return calculate_sharpe_ratio(exp_r, Rf, risk)

calculate_sharperatio("AAPL")
```

### Working with Alpha

```python
def calculate_alpha(actual_returns, betas, Rf, market_returns):
    alphas = []
    for i in range(len(actual_returns)):
        if np.isnan(actual_returns[i]) or np.isnan(betas[i]):
            alphas.append(np.nan)
        else:
            rf_decimal = Rf[i] / 100
            market_return = market_returns[i] / 100
            expected = rf_decimal + betas[i] * (market_return - rf_decimal)
            alpha = actual_returns[i]/100 - expected
            alphas.append(round(alpha * 100, 2))  # result in %
    return alphas

def calculate_alpha_for_ticker(ticker):
    # Define the path to the CSV
    path = f"/Users/tejasmacipad/Desktop/Extras/hackrush/Data/{ticker}.csv"
    
    # Get beta values
    betas_dict = calculate_beta_new(ticker, path)
    betas = list(betas_dict.values())

    # Risk-free rates and market returns
    Rf = [0.89, 1.45, 2.95, 3.88, 4.20]
    market_returns = [0.1529, 0.2879, -0.1995, 0.2473, 0.2454]  # Already in decimal form

    # Actual stock returns from the 'annual' DataFrame
    actual_returns = annual['Annual Return (%)'].values / 100  # convert from % to decimal

    # Calculate expected returns using CAPM
    expected_returns = calculate_expected_returns(betas, Rf, market_returns)

    # Calculate alpha = actual return - expected return
    alpha_values = []
    for actual, expected in zip(actual_returns, expected_returns):
        if np.isnan(actual) or np.isnan(expected):
            alpha_values.append(np.nan)
        else:
            alpha_values.append(round(actual - expected, 4))  # Keeping 4 decimal places for precision

    return alpha_values

alphas = calculate_alpha_for_ticker("AAPL")
print("Alpha values:", alphas)
```

```python
import pandas as pd
import os

def calculate_rsi(series, period=14):
    delta = series.diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)

    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# === Path to your stock CSVs ===
data_path = "/Users/tejasmacipad/Desktop/Extras/hackrush/Data"

# === List of S&P 500 tickers ===
# Make sure this list is already defined somewhere
# Example: sp500_tickers = ["AAPL", "MSFT", "GOOGL", ...]

for tick in sp500_tickers:
    if not tick:
        continue
    
    filepath = os.path.join(data_path, f"{tick}.csv")

    if(filepath == "/Users/tejasmacipad/Desktop/Extras/hackrush/Data/BRK-B.csv"):
        filepath = "/Users/tejasmacipad/Desktop/Extras/hackrush/Data/BRK.B.csv"

    if(filepath == "/Users/tejasmacipad/Desktop/Extras/hackrush/Data/BF-B.csv"):
        filepath = "/Users/tejasmacipad/Desktop/Extras/hackrush/Data/BF.B.csv"

    try:
        # Load and clean the file
        df = pd.read_csv(filepath, skiprows=2)
        df.rename(columns={df.columns[0]: "Date"}, inplace=True)
        df["Date"] = pd.to_datetime(df["Date"])
        df = df.sort_values("Date")

        # Extract and clean close price
        close_col = df.columns[1]
        df["Close"] = pd.to_numeric(df[close_col], errors='coerce')

        # Calculate indicators
        df["MA50"] = df["Close"].rolling(window=50).mean()
        df["RSI"] = calculate_rsi(df["Close"], period=14)

        # Save updated CSV
        df.to_csv(filepath, index=False)
        print(f"âœ… Updated: {tick}.csv")

    except Exception as e:
        print(f"âŒ Error with {tick}: {e}")
```

```python
import pandas as pd
import numpy as np
import os
from datetime import datetime

# === Modify this to point to your actual data folder ===
data_path = "/Users/tejasmacipad/Desktop/Extras/hackrush/Data"
tickers = [f[:-4] for f in os.listdir(data_path) if f.endswith(".csv")]

# --- Load your precomputed alpha and expected return dictionaries ---
# These should be of the form:
# alpha_dict = {"AAPL": {2020: 1.5, 2021: 0.9, ...}, ...}
# expected_return_dict = {"AAPL": {2020: 14.2, 2021: 10.3, ...}, ...}
# Load them here or build them beforehand and pass

# For example (dummy data below):
alpha_dict = {}  # Fill this with actual alpha values
expected_return_dict = {}  # Fill this with actual expected return values

def calculate_rsi(series, period=14):
    delta = series.diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)

    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# === Initialize empty DataFrame ===
date_range = pd.date_range(start="2021-01-01", end="2024-12-31", freq='D')
portfolio_matrix = pd.DataFrame(index=date_range)

# === Main loop ===
for ticker in tickers:
    try:
        file_path = os.path.join(data_path, f"{ticker}.csv")
        df = pd.read_csv(file_path, skiprows=2)
        df.rename(columns={df.columns[0]: "Date"}, inplace=True)
        df["Date"] = pd.to_datetime(df["Date"])
        df.sort_values("Date", inplace=True)
        
        close_col = df.columns[1]
        df["Close"] = pd.to_numeric(df[close_col], errors='coerce')
        df["MA50"] = df["Close"].rolling(window=50).mean()
        df["RSI"] = calculate_rsi(df["Close"])

        df.set_index("Date", inplace=True)

        filtered_df = df.loc["2021-01-01":"2024-12-31"].copy()

        marks = []

        for date, row in filtered_df.iterrows():
            year = date.year

            # Look at alpha and ER from previous year
            prev_year = year - 1
            alpha = alpha_dict.get(ticker, {}).get(prev_year, np.nan)
            exp_ret = expected_return_dict.get(ticker, {}).get(prev_year, np.nan)
            
            rsi = row.get("RSI", np.nan)
            ma50 = row.get("MA50", np.nan)
            price = row.get("Close", np.nan)

            if (
                not np.isnan(alpha) and alpha > 1 and
                not np.isnan(exp_ret) and exp_ret > 0.01 and
                not np.isnan(rsi) and rsi <= 50 and
                not np.isnan(price) and not np.isnan(ma50) and price > ma50
            ):
                marks.append((date, 1))

        # Create series for this ticker
        ticker_series = pd.Series(data=0, index=date_range)
        for d, val in marks:
            ticker_series.loc[d] = val

        portfolio_matrix[ticker] = ticker_series

        print(f"Processed: {ticker}")

    except Exception as e:
        print(f"Skipped {ticker}: {e}")

# === Save or inspect ===
portfolio_matrix.to_csv("daily_portfolio_matrix.csv")
```
